```python
hyperparameter_grid = {
    'C': [0.01, 0.1, 1, 10],
    'kernel': ['poly', 'rbf'],
    'degree': [2, 3, 4],
    'gamma':[0.001, 0.01, 0.1, 'scale', 'auto'] 
}

gscv = GridSearchCV(svr, hyperparameter_grid, scoring = 'neg_root_mean_squared_error',
                    cv = KFold(n_splits = 5, shuffle = True, random_state = 0), verbose = 2)
```
